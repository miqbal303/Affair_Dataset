### Load Required library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

### load affair dataset 

```python
fair_df = sm.datasets.fair.load_pandas().data
fair_df
```

```python
fair_df.describe()
```

```python
fair_df.info()
```

### In dataset their is no missling values all feature have datatype float64

```python
fair_df['affairs'] = (fair_df['affairs'] > 0).astype(int)
fair_df
```

```python
fair_df['rate_marriage'] = fair_df['rate_marriage'].astype(int)
fair_df['children'] = fair_df['children'].astype(int)
fair_df['religious'] = fair_df['religious'].astype(int)
fair_df['educ'] = fair_df['educ'].astype(int)
fair_df['occupation'] = fair_df['occupation'].astype(int)
fair_df['occupation_husb'] = fair_df['occupation_husb'].astype(int)
fair_df
```

### train test split

```python
X = fair_df.drop('affairs', axis = 1)
X
```

```python
y = fair_df['affairs']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

### ML algorithm Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(penalty='l2',
                        C=1.0,
                        solver='lbfgs',
                        max_iter=500,
                    
                            )
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , roc_curve, classification_report, accuracy_score
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
